# Customer Behavior and Segmentation

```python
# 1.Which customer type (Member or Normal) generates more revenue?
customer_type_revenue=df.groupby('Customer type')['Sales'].sum().reset_index()
customer_type_revenue
```

```python
#2.Is there a significant difference in spending between male and female customers?
spending_between_male_and_female_customers=df.groupby('Gender')['Sales'].sum().reset_index()
spending_between_male_and_female_customers
```

```python
#3.Which branch (Alex, Giza, etc.) serves more customers, and which generates the highest revenue?
customers_and_the_branch_they_use=df['Branch'].value_counts().reset_index()
customers_and_the_branch_they_use.columns=['Branch','customer_counts']
customers_and_the_branch_they_use

```

```python
 # step 2:revenue   from  branches
branch_revenue=df.groupby('Branch')['Sales'].sum().reset_index()
branch_revenue.columns=['Branch','Total revenue']
branch_revenue=branch_revenue.sort_values(by='Total revenue',ascending=False)
branch_revenue
```

```python
# step 3= we  are  going  to    jion  or merge  this table together
branch_Analysis=pd.merge(customers_and_the_branch_they_use,branch_revenue,on='Branch')
branch_Analysis
```

```python
#2.What is the relationship between unit price and sales volume for each product line?
#  to  do  that  we willhave  to  first   calculate   the   total sale volume and the average  unit  price 
```

```python
sales_volume=df.groupby('Product line').agg({'Quantity':sum,'Unit price':'mean'}).reset_index()
sales_volume.rename(columns={'Quantity':'Total-sales-volume','Unit price':'Avergae-unit-price'},inplace=True)
sales_volume.corr()# this  code helps  to give  a  normal understanding of  the  relationship  between Total-sales-volume and Avergae-unit-price
sales_volume

```

```python
# to visualize
custom_palette = {
    'Electronic accessories': '#1f77b4',  # blue
    'Fashion accessories': '#ff7f0e',      # orange
    'Food and beverages': '#2ca02c',       # green
    'Health and beauty': '#d62728',        # red
    'Home and lifestyle': '#9467bd',       # purple
    'Sports and travel': '#8c564b'         # brown
}
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sales_volume, x='Avergae-unit-price', y='Total-sales-volume', hue='Product line', palette=custom_palette, s=100)

plt.title='Relationship Between Unit Price and Sales Volume for Each Product Line'
plt.xlabel('Unit Price')
plt.ylabel('Sales Volume (Quantity)')
plt.legend(title='Product Line')
plt.show()
```

# From the graph, it appears that there is a positive correlation between average unit price and sales volume for most product lines. As the average unit price increases, the sales volume also tends to increase.
Electronic Accessories shows the highest sales volume at a lower average unit price, suggesting strong demand even with a lower price.
While the Health and Beauty category has a higher average unit price, it does not maintain respectable sales volume when compared to other product lines, suggesting that consumers might not be as willing to spend in this category as initially thought.
The Sports and Travel category reflects similar characteristics, with a higher price but lower sales volume, indicating potential challenges in market appeal or pricing strategies.

```python
relationship = df.groupby('Product line').apply(lambda x: x['Unit price'].corr(x['Quantity'])).reset_index()
relationship.columns=['Product line','correlationship between unit-price and quantiy']
relationship
```

```python
products = relationship['Product line']
correlation = relationship['correlationship between unit-price and quantiy']

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a bar plot
bars = plt.bar(products, correlation, color='blue', width=0.4)  # Adjust the width here

# Adding labels and title
plt.xlabel('Product Line')
plt.ylabel('Correlation between Unit Price and Quantity')
plt.title('Correlation between Unit Price and Quantity by Product Line')

# Optionally, add value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')  # va: vertical alignment
```

# Insights by Product Line
Electronic Accessories: Strong positive correlation indicates higher prices may lead to increased sales.
Fashion Accessories: Weak positive correlation suggests minimal price sensitivity; consider competitive pricing.
Food and Beverages: Low correlation indicates inelastic demand; stable pricing may be beneficial.
Health and Beauty: Similar to food and beverages; limited impact of pricing.
Home and Lifestyle: Strong correlation suggests marketing quality and value.
Sports and Travel: Low correlation; promotions might stimulate interest.
Actionable Insights
Pricing Strategies: Use premium pricing for electronic accessories and competitive pricing for fashion accessories.
Marketing Approaches: Target marketing campaigns for high-correlation categories (like home and lifestyle) and consider promotions for low-correlation categories.
Product Development: Innovate electronic accessories to justify higher prices; introduce variations in fashion accessories to attract customers.
Customer Feedback: Engage customers to understand their preferences in categories with weak correlations.
Sales Forecasting: Use correlation data for inventory management, adjusting stock levels according to price sensitivity.
Regularly revisiting this data will help adapt strategies to changing market dynamics.
